### Handling numeric data stored as text
We'll start with cleaning the numeric data that's stored as text so we can convert it to numeric type.

````python
# Price
# remove dollar sign and commas and convert to float
autos['price'] = autos['price'].str.replace('$', '').str.replace(',', '')
autos['price'] = autos['price'].astype(float)

# rename column
autos.rename(columns={'price' : 'price_usd'}, inplace=True)
print(autos.describe(include='all'))
````

````python
# Odemeter
# remove 'km' and commas and convert to int
autos['odometer'] = autos['odometer'].str.replace('km', '').str.replace(',', '')
autos['odometer'] = autos['odometer'].astype(int)

# rename column
autos.rename(columns={'odometer' : 'odometer_km'}, inplace=True)
````

### Deeper exploration into cleaned columns
Let's check the price and odometer columns to (1) verify they are formatted as expected and (2) check for any outliers.

````python
print(autos['price_usd'].unique().shape)
print(autos['price_usd'].describe())
print(autos['price_usd'].value_counts(dropna=False).head(10))
print(autos['price_usd'].value_counts(dropna=False).tail(10))
````

There are many listing with a price of 0, and it possible to have a free car listed, so we'll leave these. However, the maximum price listed is $100,000,000, which is far above the mean and even far above the third quartile. To get rid of these sorts of outliers, we'll keep only the data for which the price is within two standard deviations of the mean.

````python
mean_price_plus_2std = autos['price_usd'].mean() + 2*autos['price_usd'].std()
autos = autos[autos['price_usd'].between(0, mean_price_plus_2std)]
````

Let's look at the remaining price data.


````python
print(autos['price_usd'].unique().shape)
print(autos['price_usd'].describe())
print(autos['price_usd'].value_counts(dropna=False).head(10))
print(autos['price_usd'].value_counts(dropna=False).tail(10))
````

Now let's do a similar analysis for the odometer column.

````python
print(autos['odometer_km'].unique().shape)
print(autos['odometer_km'].describe())
print(autos['odometer_km'].value_counts(dropna=False))
````

All the values in the odometer column seem reasonable and within the same distribution. The maximum value is within two standard deviations (actually within one) of the mean. I don't see any real outliers.

Clearly there are some issues with some of the entries for registration year since the minimum year is 1000 and the maximum is 9999, neither of which are within the current modern era of vehicles.

Since a car can't have a registration date after the listing was seen, the oldest year of registration must be 2016. The mass production of cars was expanded by Henry Ford in around 1913, so we can't have any registration years prior to that. Let's remove any data points that do not fall within this range.

````python
autos = autos[autos['registration_year'].between(1913, 2016)]
print(autos['registration_year'].describe())
print(autos['registration_year'].value_counts(normalize=True))
````

## Exploring Price by Brand

````python
print(autos['brand'].value_counts())
print(autos['brand'].value_counts(normalize=True))
````

Let's analyze the brands that make up more than 5% of car listings.

````python
brand_mean_prices = {}
selected_brands = autos['brand'].value_counts().index[:7]

for b in selected_brands:
    brand_prices = autos.loc[autos['brand'] == b, 'price_usd']
    brand_mean_prices[b] = brand_prices.mean().item()
    
print(sorted(brand_mean_prices.items(), key=lambda item: item[1], reverse=True))
````

````python
brand_mean_odo = {}

for b in selected_brands:
    brand_odo = autos.loc[autos['brand'] == b, 'odometer_km']
    brand_mean_odo[b] = brand_odo.mean().item()
    
print(sorted(brand_mean_odo.items(), key=lambda item: item[1], reverse=True))
````

````python
bmp_series = pd.Series(brand_mean_prices)
bmo_series = pd.Series(brand_mean_odo)

df = pd.DataFrame(bmp_series, columns=['mean_price'])
df['mean_mileage'] = bmo_series

print(df.sort_values('mean_price', ascending=False))
print(df.sort_values('mean_mileage'))
````
